```python
import lxml.etree as ET
import sys
import os
from io import TextIOWrapper
from random import shuffle as randomsh
import matplotlib.pyplot as plt
import subprocess
import time
```

```python
def ChangeNumberOfstates(filePath: str, numberofstates_, radius_):
    tree = ET.parse(filePath)
    root = tree.getroot()
    algorithm = root.find('algorithm')
    numberofstates = algorithm.find('numberofstates')
    numberofstates.text = numberofstates_
    radius = algorithm.find('radius')
    radius.text = radius_
    tree.write(filePath, pretty_print=True)
    

def ChangeSearchType(filePath: str, searchtype_):
    tree = ET.parse(filePath)
    root = tree.getroot()
    algorithm = root.find('algorithm')
    searchtype = algorithm.find('searchtype')
    searchtype.text = searchtype_
    tree.write(filePath, pretty_print=True)
    
def ChangeToRand(filePath: str):
    tree = ET.parse(filePath)
    root = tree.getroot()
    map_data = root.find('map')
    grid = map_data.find('grid')
    for name, value in grid.items():
        print(value)

    
```

```python
#visual



from tkinter import *
from tkinter import messagebox
from tkinter import filedialog


def get_data(filePath: str, algo):
    tree = ET.parse(filePath)
    root = tree.getroot()
    log = root.find('log')
    summary = log.find('summary')
    number_of_steps = float(summary.get('numberofsteps'))
    nodescreated = float(summary.get('nodescreated'))
    length = float(summary.get('length'))
    time = float(summary.get('time'))
    return (('number of steps ' + algo , number_of_steps), ('nodescreated '+algo , nodescreated),  ('length '+algo, length), ('time '+algo, time))
        
    

def clear_labels():
    for i in range(3, 8):
        ans_label = Label(text="                 ")
        ans_label.grid(row=i, column=1, sticky="w")
        ans_label1 = Label(text="                 ")
        ans_label1.grid(row=i, column=3, sticky="w")
    return


def clear():
    name_entry.delete(0, END)
    surname_entry.delete(0, END)
    clear_labels()


def process_rstar():
    ChangeSearchType(path_map, "rstar")
    proc = subprocess.Popen([path_to_executable + " " + path_map], shell=True)
    proc.wait()
    
def process_astar():
    ChangeSearchType(path_map, "astar")
    proc1 = subprocess.Popen([path_to_executable_astar + " " + path_map], shell=True)
    proc1.wait()
    
 
def display():
    print(path_map)
    clear_labels()
    ChangeNumberOfstates(path_map, name_entry.get(), surname_entry.get())
    ChangeToRand(path_map)
    print("running astar")
    process_astar()
    print("run astar finished")
    
    BEGIN_ANS_ROW = 3
    path_map_log = path_map[:-4] + "_log.xml" 
    log_data = get_data(path_map_log, 'astar')
    for ind, value in enumerate(log_data):
        ans_label = Label(text=value[0])
        ans_label.grid(row=ind + BEGIN_ANS_ROW, column=2, sticky="w")
        ans_value = Label(text=value[1])
        ans_value.grid(row=ind + BEGIN_ANS_ROW, column=3, sticky="w")
        
    print("running rstar")
    process_rstar()
    print("run rstar finished")
    
    log_data = get_data(path_map_log, 'rstar')
    for ind, value in enumerate(log_data):
        ans_label = Label(text=value[0])
        ans_label.grid(row=ind + BEGIN_ANS_ROW, column=0, sticky="w")
        ans_value = Label(text=value[1])
        ans_value.grid(row=ind + BEGIN_ANS_ROW, column=1, sticky="w")
        
    
def get_file_name():
    global path_map
    path_map =  filedialog.askopenfilename(initialdir = "/Users/adilkhankussidenov/Documents/PathPlanning/git_rstar/Rstar_path_planning/TestData",title = "Select file",filetypes = (("jpeg files","*.xml"),("all files","*.*")))


 
root = Tk()


root.title("GUI на Python")

number_of_states = Label(root, text="Number of states:")
surname_label = Label(root, text="Radius:")


number_of_states.grid(row=0, column=0, sticky="w")
surname_label.grid(row=1, column=0, sticky="w")



name_entry = Entry()
surname_entry = Entry()

#scal = Scale(root,orient=VERTICAL ,length=300,from_=0,to=100,tickinterval=10,resolution=10)

name_entry.grid(row=0,column=1, padx=5, pady=5)
surname_entry.grid(row=1,column=1, padx=5, pady=5)

# вставка начальных данных
name_entry.insert(0, "4") 
surname_entry.insert(0, "1")

display_button = Button(text="Display", command=display)
clear_button = Button(text="Clear", command=clear)
open_file_button = Button(text="Open", command=get_file_name)

display_button.grid(row=2, column=0, padx=5, pady=5, sticky="e")
clear_button.grid(row=2, column=1, padx=5, pady=5, sticky="e")
open_file_button.grid(row=2, column=2, padx=5, pady=5, sticky="e")

root.mainloop()
```
